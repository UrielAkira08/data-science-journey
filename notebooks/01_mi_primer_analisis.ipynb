```python
# Celda 1: Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("✅ Todas las librerías importadas correctamente")

# Celda 2: Crear datos de ejemplo
datos = {
    'nombre': ['Ana', 'Luis', 'María', 'Carlos', 'Sofía'],
    'edad': [25, 30, 22, 35, 28],
    'salario': [3000, 4500, 2800, 5500, 3800]
}

df = pd.DataFrame(datos)
print(df)

# Celda 3: Análisis básico
print(f"\nEdad promedio: {df['edad'].mean():.1f} años")
print(f"Salario promedio: ${df['salario'].mean():.2f}")

# Celda 4: Tu primera visualización
plt.figure(figsize=(10, 5))
plt.bar(df['nombre'], df['salario'], color='skyblue')
plt.title('Salarios por persona')
plt.xlabel('Nombre')
plt.ylabel('Salario ($)')
plt.show()
```
